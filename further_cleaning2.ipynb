```python
import pandas as pd
import json

prop_census_isd_sorted = pd.read_csv('properties_census_isd.csv')
```

```python
print(f"Total rows in properties_census_isd.csv: {len(prop_census_isd_sorted)}")

print("Columns:")
print(prop_census_isd_sorted.columns.tolist())

#json columns
json_columns = [col for col in prop_census_isd_sorted.columns if 'json' in col.lower()]
print(f"\nJSON columns: {json_columns}")
```

```python
# Helper functions for JSON date handling
# from datetime import date, datetime 
```

```python
# Drop non-numeric columns not useful for linear regression, including JSON columns

# Identify numeric columns (int64, float64)
numeric_columns = prop_census_isd_sorted.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(f"Total numeric columns: {len(numeric_columns)}")

# Exclude JSON columns
json_cols = [col for col in prop_census_isd_sorted.columns if 'json_' in col.lower()]
print(f"JSON columns to exclude: {len(json_cols)}")

# Columns to keep: numeric columns excluding JSON
columns_to_keep = [col for col in numeric_columns if col not in json_cols]
print(f"Columns to keep: {len(columns_to_keep)}")

# Create new dataframe with only useful numeric columns
rs_numeric = prop_census_isd_sorted[columns_to_keep].copy()
#drop cols not useful for linear regression, including JSON columns
rs_numeric = rs_numeric.drop(columns=['rs_prop_sam_id', 'rs_prop_prim_sam_id', 'rs_prop_zip_code', 'rs_prop_pid', 'rs_prop_long', 'rs_prop_lat', 'rs_prop_living_area',
'rs_prop_yr_remod', 'tract', 'Housing_Tenure_owner'])

print(f"Numeric: {len(rs_numeric.columns)}")
print(rs_numeric.columns.tolist())

#drop rows with missing values in numeric columns
rs_numeric.dropna(inplace=True)

rs_numeric.head()

#export the cleaned dataframe to a new CSV file
rs_numeric.to_csv('properties_census_isd_numeric.csv', index=False)
```
