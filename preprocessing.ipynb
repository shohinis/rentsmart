# File Chopping

```python
filenames = ["1700-to-1799.csv", "1800-to-1849.csv", "1850-to-1899.csv", "1900-to-1919.csv", "1920-to-1929.csv", "1930-to-1939.csv", "1940-to-1949.csv", "1950-to-1959.csv", "1960-to-1969.csv", "1970-to-1979.csv", "1980-to-1999.csv", "2000-to-2019.csv"]
maxsize = 20000
colstokeep = ['rs_prop_sam_id', 'rs_prop_prim_sam_id', 'rs_prop_full_address',
       'rs_prop_street_number', 'rs_prop_full_street_name',
       'rs_prop_unit_number', 'rs_prop_zip_code', 'rs_prop_long', 'rs_prop_lat',
       'rs_prop_lu', 'rs_prop_owner']
for f in filenames:
    print("opening file " + f)
    properties = pd.read_csv(f)
    numrows = properties.shape[0]
    print("this has " + str(numrows) + " rows")
    dfs = [properties[i:i+maxsize] for i in range(0, len(properties), maxsize)]
    print(sum([df.shape[0] for df in dfs]))
    for i in range(len(dfs)):
        cropped = dfs[i][colstokeep]
        cropped.to_csv("chopped_rs/"+f[:-4]+"_"+str(i)+".csv")
        
```
